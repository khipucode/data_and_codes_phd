```python
# Vamos assumir que você possui um arquivo CSV chamado 'dados.csv' com pelo menos quatro colunas.
# A primeira coluna será usada como eixo X e as colunas segunda, terceira e quarta como eixo Y para o plot.

# Primeiro, precisamos importar as bibliotecas necessárias:
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Função para formatar os ticks, para número de casas decimais 1f = uma casa, 2f = duas casas
def formatar_com_virgula(x, pos):
    return f'{x:.1f}'.replace('.', ',')



# Carregar os dados do arquivo CSV em um DataFrame do pandas
#dados = pd.read_csv('https://raw.githubusercontent.com/khipucode/data_and_codes_phd/main/Respostas_em_frequ%C3%AAncia_amplificadores_RF.csv', encoding='utf-8')
df = pd.read_csv('https://raw.githubusercontent.com/khipucode/data_and_codes_phd/main/Dat_Respostas_em_frequ%C3%AAncia_amplificadores_RF.csv', delimiter=';', decimal=',')

# Verifica as primeiras linhas do DataFrame para entender a estrutura dos dados
print(df.head())

# Plotar as colunas em relação à primeira coluna (assumindo que a primeira coluna tem índice 0)
plt.figure(figsize=(10, 6))  # Define o tamanho da figura


plt.plot(df.iloc[:, 0], df.iloc[:, 1], label='$AMP_{pm}$')  # Segunda coluna
plt.plot(df.iloc[:, 0], df.iloc[:, 2], label='$AMP_{a}$')  # Terceira coluna
plt.plot(df.iloc[:, 0], df.iloc[:, 3], label='$AMP_{b}$')  # Quarta coluna
#plt.gca().xaxis.set_major_formatter(FuncFormatter(formatar_com_virgula))

# Ajustando os limites do eixo x
plt.xlim([0, 25])  # Pode ser feito também utilizando plt.gca().set_xlim([0, 10])
plt.xticks(np.arange(0, 25, 2))  # Pode ser feito também utilizando plt.gca().set_xticks(np.arange(0, 11, 1))
plt.xticks(fontsize=12)
plt.ylim([-40, 40])  # Ou plt.gca().set_ylim([-1.5, 1.5])
plt.yticks(np.arange(-40, 40, 5))
plt.yticks(fontsize=12)

plt.xlabel('Frequência (GHz)', fontsize=14)
plt.ylabel('S21 (dB)', fontsize=14)
plt.title('Gráfico de Exemplo', fontsize=16)
plt.grid(True)

plt.title('Amplificadores')  # Título do gráfico
plt.legend()  # Adiciona a legenda
plt.show()  # Exibe o gráfico
```
